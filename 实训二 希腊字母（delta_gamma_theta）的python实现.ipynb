```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']  
mpl.rcParams['axes.unicode_minus']=False   
```

```python
def delta_cl(S,K,sigma,r,T):
    import numpy as np
    from scipy.stats import norm
    d1 = (np.log(S/K)+(r+pow(sigma,2)/2)*T)/(sigma*np.sqrt(T))
    return norm.cdf(d1)
```

```python
def delta_pl(S,K,sigma,r,T):
    import numpy as np
    from scipy.stats import norm
    d1 = (np.log(S/K)+(r+pow(sigma,2)/2)*T)/(sigma*np.sqrt(T))
    return norm.cdf(d1)-1
```

```python
cl = delta_cl(S=5,K=6,sigma =0.24,r=0.04,T=0.5)
pl = delta_pl(S=5,K=6,sigma =0.24,r=0.04,T=0.5)
print(cl,pl)
```

```python
S_list = np.linspace(2.0,9.0,200)
delta_list1 = delta_cl(S=S_list,K=6,sigma =0.24,r=0.04,T=0.5)
delta_list2 = delta_pl(S=S_list,K=6,sigma =0.24,r=0.04,T=0.5)
```

```python
plt.figure(figsize=(8,6)) 
plt.plot(S_list,delta_list1,"b-",label=u'看涨期权多头',lw=2.5)
plt.plot(S_list,delta_list2 ,"r-",label=u'看跌期权空头',lw=2.5)
plt.xlabel(u"S",fontsize=13)
plt.ylabel(u'Delta',fontsize=13,rotation = 0)

plt.title(u'S与Delta的关系',fontsize=13)
plt.legend(fontsize=13)
plt.grid('True')
plt.show()
```

```python
def gamma_option(S,K,sigma,r,T):
    import numpy as np
    from scipy.stats import norm
    d1 = (np.log(S/K)+(r+pow(sigma,2)/2)*T)/(sigma*np.sqrt(T))
    return np.exp(-pow(d1,2)/2)/S*sigma*np.sqrt(2*np.pi*T)
```

```python
gamma_option(S=5,K=6,sigma =0.24,r=0.04,T=0.5)
```

```python
S_list = np.linspace(2.0,9.0,200)
gamma_list1 = gamma_option(S=S_list,K=6,sigma =0.24,r=0.04,T=0.5)
```

```python
plt.figure(figsize=(8,6)) 
plt.plot(S_list,gamma_list1,"b-",lw=2.5)
plt.xlabel(u"S",fontsize=13)
plt.ylabel(u'Gamma',fontsize=13,rotation = 0)
plt.title(u'S与Gamma的关系',fontsize=13)
plt.grid('True')
plt.show()
```

```python
T_list = np.linspace(0.1,5.0,100)
gamma_list10 = gamma_option(S=7,K=6,sigma =0.24,r=0.04,T=T_list)
gamma_list11 = gamma_option(S=6,K=6,sigma =0.24,r=0.04,T=T_list)
gamma_list12 = gamma_option(S=5,K=6,sigma =0.24,r=0.04,T=T_list)
```

```python
plt.figure(figsize=(8,6)) 
plt.plot(T_list,gamma_list10,"b-",label=u'实值看涨期权',lw=2.5)
plt.plot(T_list,gamma_list11,"r-",label=u'平值看涨期权',lw=2.5)
plt.plot(T_list,gamma_list12,"g-",label=u'虚值看涨期权',lw=2.5)
plt.xlabel(u"T",fontsize=13)
plt.ylabel(u'Gamma',fontsize=13,rotation = 0)
plt.title(u'T与Gamma的关系',fontsize=13)
plt.legend(fontsize=13)
plt.grid('True')
plt.show()
```

```python
def theta_c(S,K,sigma,r,T):
    import numpy as np
    from scipy.stats import norm
    d1 = (np.log(S/K)+(r+pow(sigma,2)/2)*T)/(sigma*np.sqrt(T))
    d2 = d1-sigma*np.sqrt(T)
    return -(S*sigma*np.exp(-pow(d1,2)/2))/2*np.sqrt(2*np.pi*T)-r*K*np.exp(-r*T)*norm.cdf(d2)
```

```python
def theta_p(S,K,sigma,r,T):
    import numpy as np
    from scipy.stats import norm
    d1 = (np.log(S/K)+(r+pow(sigma,2)/2)*T)/(sigma*np.sqrt(T))
    d2 = d1-sigma*np.sqrt(T)
    return -(S*sigma*np.exp(-pow(d1,2)/2))/2*np.sqrt(2*np.pi*T)-r*K*np.exp(-r*T)*norm.cdf(d2)+r*K*np.exp(-r*T)
```

```python
theta_p(S=5,K=6,sigma =0.24,r=0.04,T=0.5)
```
